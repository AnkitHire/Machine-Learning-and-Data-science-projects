```python
from scipy.stats import skew
data_num_skew = data_num.apply(lambda x: skew(x.dropna()))
data_num_skew = data_num_skew[data_num_skew > .75]
# apply log + 1 transformation for all numeric features with skewnes over .75
data_num[data_num_skew.index] = np.log1p(data_num[data_num_skew.index])
```

```python
#for numeric columns we will drop the null values if they are less than five percent i.e less than 260 else we will fill it with mean 
data_len = data_num.shape[0]
# check what is percentage of missing values in categorical dataframe
for col in data_num.columns.values:
 missing_values = data_num[col].isnull().sum()
 #print("{} - missing values: {} ({:0.2f}%)".format(col, missing_values, missing_values/data_len*100)) 
 
 # drop column if there is more than 50 missing values
 if missing_values > 260:
     print("droping column: {}".format(col))
     data_num = data_num.drop(col, axis = 1)
 # if there is less than 260 missing values than fill in with median valu of column
 else:
 #print("filling missing values with median in column: {}".format(col))
     data_num = data_num.fillna(data_num[col].median())
```

```python
#model preparation

#splitting in train and test

import sklearn
from sklearn.model_selection import train_test_split

```

```python
data.head()
df_x = data
df_y = y
```

```python
df_x_train , df_x_test , df_y_train , df_y_test = train_test_split(df_x , df_y , test_size = .2 , random_state=111)
print(df_x_test.shape)
print(df_x_train.shape)
print(df_y_test.shape)
print(df_y_train.shape)
```

```python
df_y.head()
```

```python
#building the model

from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(df_x_train ,df_y_train)
```

```python
pred_test = lr.predict(df_x_test)
pred_train = lr.predict(df_x_train)
```

```python
lr.intercept_
```

```python
r_sq = lr.score(df_x_train,df_y_train)  #this is the r Square
print("Rsquare:", r_sq)
```

```python
N = df_x_train.shape[0]
K= df_x_train.shape[1]
adj_rsq = 1-((1 - r_sq)* (N-1)/(N-K-1))
print("Adjusted R-Square",adj_rsq)
```

```python

error_train = df_y_train - pred_train
print("error_train",error_train,"/n/n/n")
error_test = df_y_test - pred_test
print('error_test',error_test)
```

```python
MSE = np.mean (np.square(error_test))
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
MAE = np.mean(np.abs(error_test))
MAE
```

```python
MAPE = np.mean(np.abs(error_test*100/df_y_test))
np.round(MAPE)
```

```python
Accuracy = 100 - MAPE
np.round(Accuracy)
```
