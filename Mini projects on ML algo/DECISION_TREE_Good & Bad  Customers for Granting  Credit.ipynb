### MODEL BUILDING

```python
#splitting in train and test

import sklearn
from sklearn.model_selection import train_test_split
```

```python
df_y =df.Creditability
df=df.drop(columns=['Creditability'])
```

```python
df.head()
```

```python
df_y.head()
```

```python
df_x_train , df_x_test , df_y_train , df_y_test = train_test_split(df , df_y , test_size = .2 , random_state=111)
print(df_x_test.shape)
print(df_x_train.shape)
print(df_y_test.shape)
print(df_y_train.shape)
```

```python
from sklearn.tree  import DecisionTreeClassifier
dt = DecisionTreeClassifier()# here i hv used only those values for the hyperparameters which would give us a better result

dt.fit(df_x_train , df_y_train)
```

```python
pred_test = dt.predict(df_x_test)
pred_train = dt.predict(df_x_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(df_y_test , pred_test)
```

```python
accuracy_score(df_y_train , pred_train)
```

```python
from sklearn.metrics import plot_confusion_matrix , plot_roc_curve
plot_confusion_matrix(dt , df_x_test , df_y_test)
```

```python
plot_confusion_matrix(dt , df_x_train , df_y_train)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
df_x_train_names = df_x_train.columns
fig = plt.figure(figsize=(50,30))
tree.plot_tree(dt, max_depth=2,feature_names=df_x_train_names,filled=True );
```
